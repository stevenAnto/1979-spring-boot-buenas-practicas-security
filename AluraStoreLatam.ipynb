#1. Análisis de facturación



```python

ingresos_totales = {
    'Tienda 1': tienda['Precio'].sum(),
    'Tienda 2': tienda2['Precio'].sum(),
    'Tienda 3': tienda3['Precio'].sum(),
    'Tienda 4': tienda4['Precio'].sum()
}
print(ingresos_totales)
```

```python
tineda_min_ventas = ['x',float('inf')]
ordenadoIngresosTotales = dict(sorted(ingresos_totales.items(), key=lambda item: item[1],reverse=True))
for key, value in ordenadoIngresosTotales.items():
  if value < tineda_min_ventas[1]:
    tineda_min_ventas[0]=key
    tineda_min_ventas[1]= value

  print(f'{key}: {value}')
print(f'La tineda con mas ventas es {tineda_min_ventas}')
```
